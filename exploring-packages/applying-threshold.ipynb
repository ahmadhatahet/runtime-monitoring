```python
import numpy as np
import pandas as pd
```

```python
d = pd.DataFrame(np.random.randn(5, 5))
d['y'] = np.random.randint(100, 102, (5,))
d
```

```python
col_keep = [0, 2, 3]

d[d.columns[col_keep]] = (d[d.columns[col_keep]] > 0.1).astype(np.int8) + \
                         (d[d.columns[col_keep]] > -0.1).astype(np.int8)
d
```

```python
col_delete = np.delete(d.columns[:-1], col_keep)
d.drop(col_delete, axis=1, inplace=True)

d
```

```python
def multi_thlds(x):
    x = x.reshape(x.shape[0], 1)
    x = np.where(x == [0], [0, 0], # 0
         np.where(x == [1], [1, 0], # 1
         np.where(x == [2], [0, 1], # 2
         [1, 1] ) ) )# 3
    return np.reshape(x, x.shape[0] * 2)
```

```python
np.apply_along_axis(multi_thlds, 1, d.drop('y', axis=1).to_numpy())
```
