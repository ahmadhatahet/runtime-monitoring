```python
# https://github.com/tulip-control/dd/blob/main/doc.md

from dd.autoref import BDD
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
```

```python
bdd = BDD()

bdd.declare('x1', 'x2', 'x3')

x1 = bdd.var('x1')
x2 = bdd.var('x2')
x3 = bdd.var('x3')
```

```python
f1 = x1 & ~x2 & x3
f2 = ~x1 & x2 & x3

f12 = f1 | f2
```

```python
bdd.dump('dd_test.png')
Image.open('dd_test.png')
```

```python
# test true
f3 = ~x1 & x2
(f12 & f3) == bdd.false
```

```python
f3 = ~x1 & ~x2
(f12 & f3) == bdd.false
```

# Example

```python
bdd = BDD()
vrs = ['x{i}'.format(i=i) for i in range(3)]
vrs.extend('y{i}'.format(i=i) for i in range(3))
bdd.declare(*vrs)

s = r'(x0 /\ y0) \/ (x1 /\ y1) \/ (x2 /\ y2)'  # TLA+ syntax
s = '(x0 & y0) | (x1 & y1) | (x2 & y2)'  # Promela syntax

u = bdd.add_expr(s)
bdd.incref(u)

# collect intermediate results produced while making u
bdd.collect_garbage()
```

```python
bdd.dump('dd_test.png')
Image.open('dd_test.png')
```

```python
BDD.reorder(bdd)
```
