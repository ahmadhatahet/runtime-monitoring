```python
from fastprogress import master_bar, progress_bar
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
```

#Simple check

```python
model = 'MNIST'

df_train = pd.read_csv(f'{model}_train.csv')

# convert to pattern
df_train_pattern = df_train > 0
```

```python
cls = 0

df_cls = df_train_pattern[df_train_pattern['y'] == cls].drop('y', axis=1)
print(df_cls.shape)

df_cls.drop_duplicates(inplace=True)
print(df_cls.shape)
```

```python
df_val = pd.read_csv(f'{model}_validation.csv')

# convert to pattern
df_val_pattern = df_val > 0
```

```python
val_cls = 0

total_true = 0
false_indx = []

df_val_cls = df_val_pattern[df_val_pattern['y'] == val_cls].drop('y', axis=1)

print(df_val_cls.shape)

df_val_cls.drop_duplicates(inplace=True)
print(df_val_cls.shape)

mb = master_bar(range(df_val_cls.shape[0]))

for _, (indx, row) in zip(mb, df_val_cls.iterrows()):

    pb = progress_bar(range(df_cls.shape[0]), parent=mb)
    
    for _, t in zip(pb, df_cls.iterrows()):

        c = np.all(row.values == t[1].values)    

        if c:
            false_indx.append(indx)
    
        total_true += c

    mb.main_bar.comment = str(total_true)

total_true / df_val_cls.shape[0]
```
